## Problem 7
Create a function that takes the URL of the webpage to scrape as an input, applies the code you wrote for questions 1 through 6, and generates the dataframe from question 6 as the output. [3 points]

## *Answer to Problem 7:*


```python
## Function

## Function that conducts steps 1 through 6 using URl as input, generating df as output
def scrape(url):
    
    # Create header for each scrape
    header = {"user-agent":"Nicholas Keeley (keeleyn22@darden.virginia.edu)\
    (Language = Python; Platform = MacOSX)"}

    # Parse URL HTML and save as soup object
    response=requests.get(url)
    soup=BeautifulSoup(response.text,"html5lib")
    
    # Pull titles
    stage1=soup.find_all("h3")
    titles = [t.a["title"] for t in stage1]
    
    # Pull prices
    prices=soup.find_all("p","price_color")
    prices_str=[p.string for p in prices]
    prices_final=[price.replace('Â£', '') for price in prices_str]
    
    # Pull ratings
    stars=soup.find_all("p", "star-rating")
    stars_total = [s["class"][1] for s in stars]
    
    # Pull jpeg urls
    thumbs = soup.find_all("div","image_container")
    thumbs_total=[t.a["href"] for t in thumbs]
    
    # Create DF
    d = {"Title":titles,
    "Price":prices_final,
    "Rating":stars_total,
    "Cover Art URL":thumbs_total}
    df=pd.DataFrame(d)

    return df

#scrape(url) #good
```
